### DATA PREPARATION BEFORE CLUSTER ANALYSIS

```python
categorical_cols = ['Homeowner_Pct_z4','Affordability_Index_z4','Number_of_Autos_z4',
                   'HW_BMI_z4','AIQ_Employment_z4','AIQ_ATP_v2_z4','Marriage_Code_z4','IoT_FinTech_Digital_z4','OS_RiskTaking_Health_z4',
                   'HW_ER_Visits_SC_z4','SCR_Wellness_z4','RegB_Ins_Employer_AH_z4','RegB_Ins_Market_AH_z4','HW_Stress_v2_z4',
                   'HW_Telemedicine_SC_z4','Education_Coll_v2_z4']
```

```python
data_dummies = pd.get_dummies(data2[categorical_cols], drop_first=True)
```

```python
noncategorical_cols = [col for col in data2.columns if col not in categorical_cols]
```

```python
data_preprocess = pd.concat([data2[noncategorical_cols], data_dummies], axis=1)
```

```python
data_preprocess.shape
```

```python
print(data_preprocess.isnull().sum())
```

### CLUSTER ANALYSIS BASED ON SELECTED RISK FEATURES

```python
import sklearn.cluster as cluster
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
#scaling numerical features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data_preprocess)
```

```python
sns.set(color_codes=True)
sns.set_style('darkgrid', {'axes.facecolor': '.9'})
sns.set_palette(palette='deep')
sns_c = sns.color_palette(palette='deep')
%matplotlib inline
```

```python
#This run is with scaled features. Identifying optimal clusters using below elbow-plot method
from sklearn.cluster import KMeans

inertias = []

k_candidates = range(1, 10)

for k in k_candidates:
    k_means = KMeans(random_state=42, n_clusters=k)
    k_means.fit(scaled_features)
    inertias.append(k_means.inertia_)

fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(x=k_candidates, y = inertias, s=80, ax=ax)
sns.scatterplot(x=[k_candidates[2]], y = [inertias[2]], color=sns_c[3], s=150, ax=ax)
sns.lineplot(x=k_candidates, y = inertias, alpha=0.5, ax=ax)
ax.set(title='Inertia K-Means', ylabel='inertia', xlabel='k');
```

#### Running K-means clustering using 3 clusters obtained from above plot

```python
from sklearn.cluster import KMeans
k_means = KMeans(random_state=25, n_clusters=3)
k_means.fit(scaled_features)
cluster = k_means.predict(scaled_features)
cluster = ['k-means_c_' + str(c) for c in cluster]
```

```python
data_kmeans_data=data1.assign(cluster = cluster)
```

```python
#Saving data to excel for further analysis of Risk clusters
data_kmeans_data.to_excel("kmeans_georgia_data2_092521.xlsx")
```
